```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# plot of Figure 4A

fig, ax = plt.subplots(4, 1,figsize=(12,12))

plt.subplot(411)
y = np.arange(3)
freq_XGB = [ 300,269,183]
plt.barh(y,freq_XGB, color = 'b',edgecolor = 'k')
plt.gca().invert_yaxis()
plt.yticks(y, ('O2SAT_min (300)','Age (269)', 'Encounter_type (183)'))
plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)
plt.title('XGBoost', fontweight='bold',fontsize = 20)
######
plt.subplot(412)

y = np.arange(3)
freq_LR = [300,298,128]
plt.barh(y, freq_LR, color='brown',edgecolor = 'k')
plt.gca().invert_yaxis()
plt.yticks(y, ('Age (300)','O2SAT_min (298)', 'Encounter_type (128)'))
plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)
plt.title('Logistic Regression', fontweight='bold',fontsize = 20)
#######

plt.subplot(413)

y = np.arange(3)
freq_SVM = [297,290,66]
plt.barh(y, freq_SVM, color='olive',edgecolor = 'k')
plt.gca().invert_yaxis()
plt.yticks(y, ('O2SAT_min (297)','Age (290)','COPD (66)'))
plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)
plt.title('Support Vector Machine', fontweight='bold',fontsize = 20)
#######

plt.subplot(414)

y = np.arange(3)
freq_RF = [300,225,186]
plt.barh(y,freq_RF, color='darkorange',edgecolor = 'k')
plt.gca().invert_yaxis()
plt.yticks(y, ('O2SAT_min (300)', 'Systolic_bp (225)', 'TEMP_max (186)'))
plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)
plt.title('Random Forest', fontweight='bold',fontsize = 20)
######

plt.tight_layout()

plt.savefig('Figure4A.tiff', dpi=600, format="tiff", pil_kwargs={"compression": "tiff_lzw"})
```
