```python
import itertools
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Distribution plot of covid-19 patients AGE & O2SAT_min
# Read following data files
data_age_alive = pd.read_csv('alive_patients_age.csv')
data_age_deceased = pd.read_csv('deceased_patients_age.csv')
data_o2sat_min_alive = pd.read_csv('alive_patients_o2sat_min.csv')
data_o2sat_min_deceased = pd.read_csv('deceased_patients_o2sat_min.csv')
```

```python
# Figure 4B (Violin plot of COVID_19 patient's O2SAT_min (deceased & alive))

fig, (ax1, ax2) = plt.subplots(ncols=2,figsize=(8,4))

sns.violinplot(x='O2SAT_min',data=data_o2sat_min_deceased, ax=ax1, color="gold",orient='v', medianprops={"zorder":3})
sns.violinplot(x='O2SAT_min',data=data_o2sat_min_alive, ax=ax2, color="gold",orient='v', medianprops={"zorder":3})

ax1.set_title('Deceased (n=391)',fontsize=18)
ax1.set_ylabel('O2SAT_min (%)' ,fontsize=18)
ax1.yaxis.set_tick_params(labelsize=16)
ax1.set_ylim(-20, 120)

ax2.set_title('Alive (n=4411)',fontsize=18)
ax2.set_ylabel('O2SAT_min (%)' ,fontsize=18)
ax2.yaxis.set_tick_params(labelsize=16)
ax2.set_ylim(-20, 120)

plt.tight_layout()

# save figure
plt.savefig("Figure_4B.tiff", dpi=600, format="tiff", pil_kwargs={"compression": "tiff_lzw"})
```

```python
# Figure 4C (Violin plot of COVID_19 patient's Age (deceased & alive))

fig, (ax1, ax2) = plt.subplots(ncols=2,figsize=(8,4))

sns.violinplot(x='Age',data=data_age_deceased, ax=ax1, color="gold",orient='v', medianprops={"zorder":3})
sns.violinplot(x='Age',data=data_age_alive, ax=ax2, color="gold",orient='v', medianprops={"zorder":3})

ax1.set_title('Deceased (n=391)',fontsize=18)
ax1.set_ylabel('Age (years)' ,fontsize=18)
ax1.yaxis.set_tick_params(labelsize=16)
ax1.set_ylim(0, 100)

ax2.set_title('Alive (n=4411)',fontsize=18)
ax2.set_ylabel('Age (years)' ,fontsize=18)
ax2.yaxis.set_tick_params(labelsize=16)
ax2.set_ylim(0, 100)

plt.tight_layout()

# save figure
plt.savefig("Figure_4C.tiff", dpi=600, format="tiff", pil_kwargs={"compression": "tiff_lzw"})
```
